**Lets Check Outliers**

```python
feature_cols = wine.columns.drop('quality')

# Define the grid structure for the subplots
n_features = len(feature_cols) # 11 features
n_rows = 3
n_cols = 4 # 3 rows * 4 columns = 12 subplots, which is enough for 11 features

plt.figure(figsize=(18, 12))
plt.suptitle('Box Plots for Red Wine Quality Features (Outlier Detection)', fontsize=16, y=1.02)

# Loop through each feature column and create a box plot in a subplot
for i, col in enumerate(feature_cols):
    plt.subplot(n_rows, n_cols, i + 1) 
    
    sns.boxplot(y=wine[col], color='skyblue')
    
    # Set titles and clean up labels
    plt.title(col, fontsize=12)
    plt.ylabel('')
    plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

**This boxplots confirm that outliers are present in every column and because Linear Regression is Sensitive to Outliers,it is recommended to Treat them** 

```python
#We will use CAPPING 
for col in feature_cols:
    Q1 = wine[col].quantile(0.25)
    Q3 = wine[col].quantile(0.75)
    IQR = Q3-Q1

    Upperlimit = Q3 + (1.5*IQR)
    Lowerlimit = Q1 - (1.5*IQR)

    wine[col] = np.where(wine[col] > Upperlimit, Upperlimit, wine[col])
    # Replace values below the lower bound with the lower bound
    wine[col] = np.where(wine[col] < Lowerlimit, Lowerlimit, wine[col])
```

```python
##Printing Descriptive Statistics 
print(wine[feature_cols].describe().to_markdown(numalign="left", stralign="left"))
```

```python
# Lets check if the Outliers have been removed or not
plt.figure(figsize=(18, 12))
plt.suptitle('Box Plots for Red Wine Quality Features (Outlier Detection)', fontsize=16, y=1.02)

# Loop through each feature column and create a box plot in a subplot
for i, col in enumerate(feature_cols):
    plt.subplot(n_rows, n_cols, i + 1) 
    sns.boxplot(y=wine[col], color='skyblue')
    plt.title(col, fontsize=12)
    plt.ylabel('')
    plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

**In the Boxplots Above we can see all the outliers have been removed,Now we can move to the next Problem**

# Model Creation

```python
# Independent and Dependent Variable
X = wine.drop('quality',axis=1)
Y = wine['quality']
```

```python
X
```

```python
Y
```

```python
# Training and Testing Data
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
len(Y_train)
```

```python
len(Y_test)
```

```python
# Creating a Model
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,Y_train)
```

```python
Y_predict = LR.predict(X_test)
```

```python
Y_predict # Predicted Values
```

```python
Y_test # Actual Values
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
r2 = r2_score(Y_test,Y_predict)
```

```python
r2
```

**An R2 Score of 0.41 is considered moderate for this type of problem. It tells you that a simple Linear Regression model, while able to identify important factors, is not highly accurate at predicting the wine quality score**
