Above table shows there are 4 values of ? , So we will have to change it 

```python
# Importing this to avoid any warnings
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# So we will replace ? with NaN
df['Age'].replace('?', np.nan, inplace=True)
```

```python
df['Age'].value_counts()
```

So in the above table you can see ? has been replaced

```python
# Converting Age column into float dtype 
df['Age'] = pd.to_numeric(df['Age'])
```

```python
df['Age'].info()
```

```python
# Now the same method with the Smoking column
df['Smoking'].value_counts()
```

```python
df['Smoking'].replace('?', np.nan, inplace=True)
```

```python
df['Smoking'].value_counts()
```

```python
# Converting Smoking column from Object dtype to float
df['Smoking'] = pd.to_numeric(df['Smoking'])
```

```python
df['Smoking'].info()
```

```python
# Here we can see All the values are in Numeric dtype 
df.info()
```

```python
# As we replaced The ? with Nan , there are many missing values so we will have to impute them using Median
```

```python
median_value = df['Age'].median()
```

```python
median_value
```

```python
df['Age'].fillna(median_value, inplace=True)
```

```python
median_value1 = df['Smoking'].median()
```

```python
median_value1
```

```python
df['Smoking'].fillna(median_value1, inplace=True)
```

# Model Creation


```python
# Step 1 -Independent and Dependent Variable Creation
X = df[['Age','gender','Body_mass_index','No_of_children','Smoking','Region']]
Y = df[['Insurance_price']]
```

```python
X
```

```python
Y
```

```python
# Step 2 - Training and Testing data 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state = 4)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
len(Y_train)
```

```python
len(Y_test)
```

```python
# Step 3 - Creating a Model
from sklearn.linear_model import LinearRegression 
LR = LinearRegression()
LR.fit(X_train,Y_train)
```

```python
Y_predict=LR.predict(X_test)
```

```python
Y_predict # Predicted Values
```

```python
Y_test # Actual Values
```

```python
# Step 4 - Model Evaluation 
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
r2 = r2_score(Y_test,Y_predict)
```

```python
r2
```

**This means the Predicted Values by the model are 73 % Accurate**
