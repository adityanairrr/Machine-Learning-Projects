```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder,OrdinalEncoder
import warnings
import matplotlib.pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import FunctionTransformer
import pickle
from sklearn.pipeline import Pipeline
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('obesity_level.csv')
```

```python
data
```

```python
data.info()
```

```python
data = data.drop("id",axis=1)
```

```python
data.columns
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['0be1dad'].value_counts()
```

```python
data = data.rename(columns={
    'family_history_with_overweight': 'Familyoverweight_history',
    'FAVC': 'HighCaloriefood_consumption',
    'FCVC': 'Consumption_of_vegetables',
    'NCP': 'No_of_mainmeals',
    'CAEC': 'food_between_meals',
    'SMOKE': 'Smoking',
    'CH20': 'WaterConsumption',
    'SCC': 'Beverages_consumption',
    'FAF': 'Physical_Activity',
    'TUE': 'Time_on_devices',
    'CALC': 'Alcohol',
    'MTRANS': 'Mode_of_transportation',
    '0be1dad': 'Result'})
    
```

```python
data['Result'] = data['Result'].replace({'0rmal_Weight': 'Normal_Weight'})
data['Mode_of_transportation'] = data['Mode_of_transportation'].replace({'Public_Transportation': 'Public'})
```

```python
data.columns
```

```python
data
```

```python
data['Result'].value_counts()
```

```python
data.info()
```

```python
le = LabelEncoder()
cat_var = ['Gender','food_between_meals','Alcohol','Mode_of_transportation']
```

```python
for col in cat_var:
    data[col] = le.fit_transform(data[col])
```

```python
data
```

Transportation 
Automobile-0 , Bike-1, Motorbike-2 , Public-3 , Walking-4

```python
data['Alcohol'].value_counts()
```

```python
x = data.drop("Result",axis=1)
y= data["Result"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.2,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,f1_score,classification_report
```

```python
y_pred = dt.predict(x_train)
accuracy_score(y_train,y_pred)
```

```python
y_predict = dt.predict(x_test)
accuracy_score(y_test,y_predict)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'criterion':['gini','entropy'],'max_depth':[5,8,10,12,20,30],'min_samples_split':[5,10,15,20]}
```

```python
grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=params,refit=True,
    cv=5, 
    scoring='accuracy',
    n_jobs=-1 )# Use all available CPU cores
```

```python
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
from sklearn.tree import plot_tree

dt_diagram = DecisionTreeClassifier(max_depth=3, criterion='gini', random_state=42)
dt_diagram.fit(x, y)

plt.figure(figsize=(65, 50))
plot_tree(dt_diagram, 
          feature_names=x.columns, 
          class_names=dt.classes_, 
          filled=True, 
          rounded=True,fontsize=22)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_train)
accuracy_score(y_train,y_pred)
```

```python
y_predict = rf.predict(x_test)
accuracy_score(y_test,y_predict)
```

```python
rf_balanced = RandomForestClassifier(n_estimators=200,max_depth=12,min_samples_leaf=15,random_state=42,min_samples_split=40)
rf_balanced.fit(x_train,y_train)
```

```python
y_pred1 = rf_balanced.predict(x_train)
accuracy_score(y_train,y_pred1)
```

```python
y_predict1 = rf_balanced.predict(x_test)
accuracy_score(y_test,y_predict1)
```

```python
import pickle
```

```python
file_name = "Obesity_check_model.pkl"

with open('Obesity_check_model.pkl', 'wb') as file:
    pickle.dump(rf_balanced, file)

with open('label_encoder.pkl', 'wb') as file:
    pickle.dump(le, file)

feature_columns = x_train.columns.tolist()
with open('feature_columns.pkl', 'wb') as file:
    pickle.dump(feature_columns, file)
```
